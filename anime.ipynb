```python
# convert 'genre' column from object to list
def clean_genre(genres):
    genres = [item.strip('][').split(', ') for item in genres]
    
    # remove quote from each element
    genres = [[item.replace("'", '') for item in word] for word in genres]
    
    genres = ['|'.join(item) for item in genres]
    return genres
```
